## b) Linear model estimation


I am building a linear model that can be represented as follows:

$$\hat{y} = \beta*X + \epsilon$$


and such a model is interpreted as follows: An increase in xi by a unit causes an increase in y by Î²i ceteris paribus
units


```python
X, y = sales_numeric_df.loc[:, "Brand1_SVAL":"WeekNumeric"], sales_numeric_df.Brand1_SVOL
```

```python
from sklearn import linear_model
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

model = linear_model.LinearRegression()
```

```python
model.fit(X, y)
```

```python
y_model = model.predict(X)
```

### c) Model assesment

```python

#  numbers are far away from zero, therefore a map can be defined
def mean_absolute_percentage_error(y_pred, y): 
    y, y_pred = np.array(y), np.array(y_pred)
    return np.mean(np.abs((y_pred - y) / y)) * 100
```

```python

r2 = r2_score(y_model, y)

mse = mean_squared_error(y_model, y)
rmse = np.sqrt(mse)

mae = mean_absolute_error(y_model, y)
mape = mean_absolute_percentage_error(y_model, y)
```

```python
pd.DataFrame([("Intercept", model.intercept_)]+[(var, coeff) for var, coeff in zip(X.columns, model.coef_)], columns = ["Variable name", "Coefficient"])
```

```python

print(f"R2: {round(r2, 4)}\n")
print(f"MSE: {round(mse, 3)}")
print(f"RMSE: {round(rmse, 3)}\n")
print(f"MAE: {round(mae, 3)}")
print(f"MAPE: {round(mape, 3)}")

```

---

The R2 coefficient shows how well the model fits the data. It indicates how much the variance of proposed model is smaller than the variance of of the mean.

The remaining metrics give a better look at the results, e.g. the calculated mape value shows that the inaccuracy of the created model is ~ 4% on average

---

```python
# wspÃ³Å‚czynnik autokorelacja
epsilon = np.array(y_model) - np.array(y)
autocorr = pd.DataFrame(epsilon)[0].autocorr()
print(f"Autocorrelation: {autocorr}")
```

---

Autocorrelation is low for residuals, so residuals are distributed randomly - it satisfies the assumptions of linear regression model

---

### d) Plots: Residuals and Actual vs Fitted

```python
# residual chart
epsilon = np.array(y_model) - np.array(y)
eps_mean, eps_std = np.mean(epsilon), np.std(epsilon)

fig, ax = plt.subplots(figsize=(30,8))
ax.plot(range(len(epsilon)),epsilon, 'o', color='black', alpha=.3)
ax.axhline(eps_mean, color='#dd3344', alpha=.5)
ax.axhline(eps_mean - 3*eps_std, color='#99aa44', linestyle='dashed')
ax.axhline(eps_mean + 3*eps_std, color='#99aa44', linestyle='dashed')


ax.set_ylabel("epsilon")
plt.show()
```

---

Based on the residual diagram and the high R2 coefficient, it can be concluded that heteroscedasticity does not occur. No need to use the test (Breusch-Pagan)
